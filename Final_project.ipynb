# Overview

- This notebook includes the final project implementation for DATS 6202_10 Machine Learning I

- Here we will work on kaggle competation [Bike Sharing Demand
Forecast use of a city bikeshare system](https://www.kaggle.com/competitions/bike-sharing-demand/overview).

- In this competition, we are provided with two years of hourly bike rental data. The dataset is split such that the first 19 days of each month form the training set, and data from the 20th to the end of the month constitute the test set.

- Our challenge is to accurately predict the total number of bikes rented during each hour of the test set. Crucially, predictions must be made using only the information available before the start of each rental hour, ensuring the model's predictive capability is tested under realistic conditions of prior knowledge.




 # Introduction


 - "Bike Sharing Demand Forecast" competition hosted by Kaggle, utilizing real-time data from Capital Bikeshare. Our approach includes rigorous data preprocessing and feature engineering techniques such as cyclical encoding for time-based features and feature scaling to normalize data distributions. We have trained and fine-tuned Random Forest, XGBoost, and CatBoost Regression models, focusing on optimizing hyperparameters to minimize the Root Mean Squared Logarithmic Error (RMSLE), which is the primary evaluation metric for this competition. Based on the models with the lowest RMSLE scores, we predict the count of bike rentals for each datetime in the test dataset and prepare the submission file, which is saved in the directory.

# Description of the Dataset

- **datetime** - hourly date + timestamp  

- **season** -  1 = spring, 2 = summer, 3 = fall, 4 = winter

- **holiday** - whether the day is considered a holiday

- **workingday** - whether the day is neither a weekend nor holiday

- **weather** - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
              2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
              3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
              4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

- **temp**- temperature in Celsius

- **atemp** - "feels like" temperature in Celsius

- **humidity** - relative humidity

- **windspeed** - wind speed

- **casual** - number of non-registered user rentals initiated

- **registered** - number of registered user rentals initiated

- **count** - number of total rentals - **(Target)**

#Feature Engineering

##Cyclical encoding Train and Test data

```python

df_train['datetime'] = pd.to_datetime(df_train['datetime'])
df_train['year'] = df_train['datetime'].dt.year
df_train['month'] = df_train['datetime'].dt.month
df_train['day'] = df_train['datetime'].dt.day
df_train['hour'] = df_train['datetime'].dt.hour
df_train['dayofweek'] = df_train['datetime'].dt.dayofweek

# Cyclical encoding for hour and month

df_train['hour_sin'] = np.sin(df_train['hour'] * (2. * np.pi / 24))
df_train['hour_cos'] = np.cos(df_train['hour'] * (2. * np.pi / 24))
df_train['month_sin'] = np.sin((df_train['month']-1) * (2. * np.pi / 12))
df_train['month_cos'] = np.cos((df_train['month']-1) * (2. * np.pi / 12))

# Drop the original datetime column
df_train.drop('datetime', axis=1, inplace=True)

```

```python
#Cyclical encoding for test data

df_test['datetime'] = pd.to_datetime(df_test['datetime'])
df_test['year'] = df_test['datetime'].dt.year
df_test['month'] = df_test['datetime'].dt.month
df_test['day'] = df_test['datetime'].dt.day
df_test['hour'] = df_test['datetime'].dt.hour
df_test['dayofweek'] = df_test['datetime'].dt.dayofweek


df_test['hour_sin'] = np.sin(df_test['hour'] * (2. * np.pi / 24))
df_test['hour_cos'] = np.cos(df_test['hour'] * (2. * np.pi / 24))
df_test['month_sin'] = np.sin((df_test['month']-1) * (2. * np.pi / 12))
df_test['month_cos'] = np.cos((df_test['month']-1) * (2. * np.pi / 12))


df_test.drop('datetime', axis=1, inplace=True)
```

## Imputing missing values

```python
from sklearn.impute import SimpleImputer

if len(df_miss['var']) > 0:
    si = SimpleImputer(missing_values=np.nan, strategy='mean')

    # Imputing the variables with missing values in df_train, df_val and df_test
    df_train[df_miss['var']] = si.fit_transform(df_train[df_miss['var']])
    df_val[df_miss['var']] = si.transform(df_val[df_miss['var']])
    df_test[df_miss['var']] = si.transform(df_test[df_miss['var']])
```

## Splitting the feature and target

```python

# feature matrix as DataFrame
X_train = df_train[np.setdiff1d(df_train.columns, [target])]
X_val = df_val[np.setdiff1d(df_val.columns, [target])]
X_test = df_test[np.setdiff1d(df_test.columns, [target])]

# target vector
y_train = df_train[target].values
y_val = df_val[target].values
y_test = df_test[target].values




```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
# List of features to scale
features_to_scale = ['temp', 'atemp', 'humidity', 'windspeed']

X_train[features_to_scale] = ss.fit_transform(X_train[features_to_scale])

X_val[features_to_scale] = ss.transform(X_val[features_to_scale])

X_test[features_to_scale] = ss.transform(X_test[features_to_scale])

X_train = X_train.values
X_val = X_val.values
X_test = X_test.values

```

```python
# Standardizing the training data
y_train = ss.fit_transform(y_train.reshape(-1, 1)).reshape(-1)

# Standardizing the validation data
y_val = ss.transform(y_val.reshape(-1, 1)).reshape(-1)

# Standardizing the test data
y_test = ss.transform(y_test.reshape(-1, 1)).reshape(-1)
```

#Hyperparameter Tuning

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_log_error
import numpy as np

rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

#RMSLE
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))


rmsle_scorer = make_scorer(rmsle, greater_is_better=False)


grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid, scoring=rmsle_scorer, cv=2, verbose=2, n_jobs=-1)


grid_search_rf.fit(X_train, y_train)


print("Best parameters found: ", grid_search_rf.best_params_)



y_pred = grid_search_rf.best_estimator_.predict(X_test)


y_pred_clipped = np.clip(y_pred, a_min=0, a_max=None)


rmsle_rf = rmsle(y_test, y_pred_clipped)


print(f'RMSLE: {rmsle_rf}')

# The list to hold results from all models
params_estimator_gs = []

# Append Random Forest results
params_estimator_gs.append([
    grid_search_rf.best_params_,
    grid_search_rf.best_estimator_,
    rmsle_rf
])



```

#XGBOOST

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_log_error
import numpy as np


xgb_model=XGBRegressor(random_state=42)

param_grid = {
    'colsample_bytree': [0.2 ,0.3,0.35],
    'learning_rate': [0.009,0.01],
    'max_depth': [5, 5.5],
    'alpha': [10, 11],
    'n_estimators': [90,100]
}


grid_search_xg = GridSearchCV(estimator=xgb.XGBRegressor(objective='reg:squarederror'),
                           param_grid=param_grid, scoring='neg_mean_squared_error',
                           cv=3, verbose=1)
grid_search_xg.fit(X_train, y_train)


print("Best parameters:", grid_search_xg.best_params_)


y_pred = grid_search_xg.predict(X_test)
y_pred_clipped = np.clip(y_pred, a_min=0, a_max=None)
rmsle_xg = rmsle(y_test, y_pred_clipped)


print(f'RMSLE: {rmsle_xg}')


# Append  results
params_estimator_gs.append([

    grid_search_xg.best_params_,
    grid_search_xg.best_estimator_,
    rmsle_xg
])


```
