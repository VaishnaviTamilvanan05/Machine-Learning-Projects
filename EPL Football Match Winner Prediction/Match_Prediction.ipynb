```python
matches["date"] = pd.to_datetime(matches["date"])
matches.dtypes
```

```python
#creating predictors for ML
matches["venue_code"] = matches["venue"].astype("category").cat.codes #convert string to categorial data and then to integers
matches["opp_code"] = matches["opponent"].astype("category").cat.codes #give each team a code
matches["hour"] = matches["time"].str.replace(":.+","", regex=True).astype("int") #regular expression replace : with ""
matches["day_code"] = matches["date"].dt.dayofweek
matches["target"] = (matches["result"] == "W").astype("int") #W~True and L&D ~ false
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#subset
train = matches[matches["date"] <'2023-01-01']
test = matches[matches["date"] >'2023-01-01']
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors],train["target"]) #fit to train predictors and predict target
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"], preds)
acc
```

```python
#create DF that will combine actual(target) and predicted values(preds)
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
```

```python
from sklearn.metrics import precision_score
precision_score(test["target"],preds) 
```

```python
#Improving precison with rolling averages
grouped_matches = matches.groupby("team") #split matches df and group by team(i.e, one df for each team)
group = grouped_matches.get_group("Manchester United")
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3,closed='left').mean() #to exclude current week closed =left
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #remove rows with missing values
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols =[f"{c}_rolling" for c in cols]
new_cols         
```

```python
rolling_averages(group, cols, new_cols)
```

```python
#matches df group by team and one df for each team
#apply rolling_avaerages function to each team
matches_rolling =matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team') #drop extra index levels
matches_rolling.index = range(matches_rolling.shape[0]) #unique value for index
matches_rolling
```

```python
#Retrain ML model

def make_predictions(data, predictors):
    train = data[data["date"] <'2023-01-01']
    test = data[data["date"] >'2023-01-01']
    rf.fit(train[predictors],train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds), index=test.index)
    precision = precision_score(test["target"],preds) 
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
```
