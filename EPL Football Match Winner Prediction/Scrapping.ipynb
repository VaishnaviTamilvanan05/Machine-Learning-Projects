```python
#Scraping data for multiple season and team with a loop
```

```python
years = list(range(2023, 2020, -1))
years
```

```python
all_matches =[] #initialize list that will contain several DF, each DF will contain match log of one team and season
```

```python
standings_url ="https://fbref.com/en/comps/9/Premier-League-Stats" #Url for scrapping
```

```python
import time
for year in years:
    data = requests.get(standings_url) #download html
    soup = BeautifulSoup(data.text) #parse html
    standings_table = soup.select('table.stats_table')[0] #select table which contains individual team url
    
    links = [l.get("href") for l in standings_table.find_all('a')] #get all teams link
    links = [l for l in links if '/squads/' in l] #get only squad links
    team_urls = [f"https://fbref.com{l}" for l in links] #convert to absolute urls
    
    previous_season = soup.select("a.prev")[0].get("href") #get relative prev url
    standings_url = f"https://fbref.com/{previous_season}" #absolute url
    
    
    for team_url in team_urls: #individually scrap match logs for each team
        team_name = team_url.split("/")[-1].replace("-Stats","").replace("-"," ") #format team name
        
        data = requests.get(team_url) #download html
        soup = BeautifulSoup(data.text) #parse html
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0] #select from table with matching string
        
        soup = BeautifulSoup(data.text) #parse
        links = [l.get("href") for l in soup.find_all('a')] #getting shooting stats link
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f"https://fbref.com{links[0]}") #to download html of shooting stats page
        shooting = pd.read_html(data.text , match="Shooting")[0] #using pandas to read data.text, match string and take first element in returned list
        shooting.columns = shooting.columns.droplevel() #drop one index in multi level index
        time.sleep(5)
        try:
            team_data = matches.merge(shooting[["Date","Sh","SoT","Dist","FK","PK","PKatt"]], on="Date") #merge matches and shooting stats
        except ValueError:
            continue
            
        team_data = team_data[team_data["Comp"] == "Premier League"] #Select on PL competition
        team_data["Season"] = year #new colum to identify
        team_data["Team"] = team_name
        all_matches.append(team_data) #add all data to list
        time.sleep(5) #to avoid getting locked by website while scrapping rapidly and slowing down website by quick requests
            

       
```

```python
match_df = pd.concat(all_matches) #concats multiple lists and return one data frame
match_df.columns =[c.lower() for c in match_df.columns]
```

```python
match_df.to_csv("PLmatches.csv") #write in csv file using pandas
```
